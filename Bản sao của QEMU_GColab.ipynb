```python
#@title QEMU Virtual Machine
import os

NGROK_TOKEN =""#@param {type:"string"}
CDROM =""#@param {type:"string"}

if NGROK_TOKEN =="":
 print("Please enter your token")
elif len (str(NGROK_TOKEN)) < 49:
 print("Your ngrok token is invalid.Please check")
elif len(str(NGROK_TOKEN)) > 49:
 print("Your ngrok token is invalid. Please check.")
elif CDROM == "":
 print("Please put your iso")
else:
 os.system("wget https://bim.equinox.io/c/4VmDzA7iaHb/ngrok-stable-linux-amd64.zip")
 os.system("unzip -o ngrok-stable-linux-amd64.zip")
 os.system("rm -rf ngrok-stable-linux")
 os.system(f"./ngrok authtoken {NGROK_TOKEN}")
 os.system("nohup ./ngrok tcp --region ap 5900 &>/dev/null &")
 print("Download Driver Qemu...")
 os.system('wget -O virtio-win.iso https://fedorapeople.org/groups/virt/virtio-win/direct-downloads/archive-virtio/virtio-win-0.1.215-1/virtio-win-0.1.215.iso')
 print("Download Your Windows...")
 os.system(f'wget -O CDROM.iso {CDROM}')
 print("Install Qemu...")
 os.system("sudo apt-get update")
 os.system("apt-get install qemu -y")
 os.system("sudo apt install qemu-utils -y")
 os.system('sudo apt install qemu-system-x86 -y')
 os.system('qemu-img create -f raw HDD.img 1T')
 print("Your Vnc Ip. Enjoy!:")
!wget -O run.sh https://raw.githubusercontent.com/QuocNinhXR/QEMU_GColab/main/run.sh
!chmod +x run.sh
!./run.sh
```
